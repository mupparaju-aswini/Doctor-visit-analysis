# visualize is there is any missing values in the dataset based on a heat map

```python
sns.heatmap(d.isnull(),cbar=False,cmap='viridis')
```

```python
non_numeric_columns = d.select_dtypes(exclude=[np.number]).columns
print(non_numeric_columns)
```

```python
d_numeric = d.drop(non_numeric_columns, axis=1)
```

```python
d['gender'] = d['gender'].map({'female': 1, 'male': 0})
d['private'] = d['private'].map({'yes': 1, 'no': 0})
d['freepoor'] = d['freepoor'].map({'yes': 1, 'no': 0})
d['freerepat'] = d['freerepat'].map({'yes': 1, 'no': 0})
d['nchronic'] = d['nchronic'].map({'yes': 1, 'no': 0})
d['lchronic'] = d['lchronic'].map({'yes': 1, 'no': 0})
```

# Find out the correlation between variables in the given dataset correlation between variables

```python
plt.figure(figsize=(10,10))
sns.heatmap(d.corr(),cbar=True,annot=True,cmap='Blues')
```

# Analyze how the income of a patient affects the no of visits to the hospital

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=d)
plt.xlabel('income')
plt.ylabel('visits')
```

# Count and visualize the number of males and females affected by illness

```python
sns.histplot(d.gender,bins=2)
```

# visualize the percentage of people getting govt health insurance due to low income due to old age and also the percentage of people having private health

```python
label=[1,0]
Y=d[d['freepoor']==1]
N=d[d['freepoor']==0]
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting health insurance due to low income")
plt.show()
Y=d[d['private']==1]
N=d[d['private']==0]
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people have private insurance")
plt.show()
Y=d[d['freerepat']==1]
N=d[d['freerepat']==0]
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting health insurance due to old age,disability or veteran status")
plt.show()
```

# plot a horizontal barchart to analyze the reduced days of activity due to illness based gender

```python
db= d.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
plt.title('Barchart')
plt.xlabel('reduced activity')
plt.ylabel('gender')
plt.show()
```
